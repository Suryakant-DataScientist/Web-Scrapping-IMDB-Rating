```python
soup = BeautifulSoup(source.text,'html.parser')
```

```python
#print(soup)
```

```python
movies = soup.find('tbody', class_="lister-list")
```

```python
#print(movies)
```

```python
movies = soup.find('tbody', class_="lister-list").find_all('tr')
```

```python
print(len(movies))
```

```python
# Name
# movies.find('td',class_="titleColumn").a.text
```

```python
# Rank
```

```python
# Year
# movies.find('td',class_="titleColumn").span.text.strip('()')
```

```python
# Rating
# movies.find('td',class_="ratingColumn imdbRating").strong.text
```

```python
for movie in movies:
    name = movie.find('td',class_="titleColumn").a.text
    rank = movie.find('td',class_="titleColumn").get_text(strip = True).split('.')[0]
    rating = movie.find('td',class_="ratingColumn imdbRating").strong.text
    year = movie.find('td',class_="titleColumn").span.text.strip('()')
    print(rank,name,year,rating)
    
```

### How to save these in excel file

```python
excel = openpyxl.Workbook()
```

```python
# check the sheets available in the workbook
print(excel.sheetnames)
```

```python
sheet = excel.active
```

```python
sheet.title = 'Top Rated Movies'
```

```python
print(excel.sheetnames)
```

```python
sheet.append(['Movie Rank','Movie Name','Year Of Release','IMDB Rating'])
```

```python
for movie in movies:
    name = movie.find('td',class_="titleColumn").a.text
    rank = movie.find('td',class_="titleColumn").get_text(strip = True).split('.')[0]
    rating = movie.find('td',class_="ratingColumn imdbRating").strong.text
    year = movie.find('td',class_="titleColumn").span.text.strip('()')
    print(rank,name,year,rating)
    sheet.append([rank,name,year,rating])
```
